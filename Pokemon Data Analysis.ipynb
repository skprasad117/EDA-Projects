## 5. Get all Pokemons whose spawn time is less than 300 seconds.

```python
# imputing Nan with 0:0
df['spawn_time'] =df['spawn_time'].fillna('0:0')
```

```python
# new feature for spawn time is created in seconds as unit
for index, row in df.iterrows():
    updated_value = int(row['spawn_time'].split(":")[0])*60 + int(row['spawn_time'].split(":")[1])
    df.at[index,'spawn_time_in_sec'] = updated_value
```
